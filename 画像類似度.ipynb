```
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
from skimage import data, img_as_float
from skimage.measure import compare_ssim as ssim
from skimage.color import rgb2gray
```

```
image1 = plt.imread('/content/images/44041162_CR_20100510_001.jpg')
image1 = rgb2gray(image1)
img1 = img_as_float(image1)

image2 = plt.imread('/content/images/44041162_CR_20110801_001.jpg')
image2 = rgb2gray(image2)
img2 = img_as_float(image2)
```

```
image3 = plt.imread('/content/images/44041162_CR_20190409_001.jpg')
image3 = rgb2gray(image3)
img3 = img_as_float(image3)
```

```
def mse(x, y):
    return np.linalg.norm(x - y)
```

```
mse_none = mse(img1, img1)
ssim_none = ssim(img1, img1, data_range=img1.max() - img1.min())
```

```
mse_2009 = mse(img1, img2)
ssim_2009 = ssim(img1, img2,
                  data_range=img2.max() - img2.min())
```

```
mse_2019 = mse(img1, img3)
ssim_2019 = ssim(img1, img3,
                  data_range=img3.max() - img3.min())
```

```
test_ssim = compare_ssim(img1, img3, data_range=img3.max() - img3.min())
print(test_ssim)
```

```
fig = plt.figure(figsize=(10,4))
grid = ImageGrid(fig, 111,  
                 nrows_ncols=(1, 3),  
                 axes_pad=0.1)

label = 'MSE: {:.2f}, SSIM: {:.2f}'

grid[0].imshow(img1, cmap=plt.cm.gray, vmin=0, vmax=1)
grid[0].set_xlabel(label.format(mse_none, ssim_none))
grid[0].set_title('Original image')

grid[1].imshow(img2, cmap=plt.cm.gray, vmin=0, vmax=1)
grid[1].set_xlabel(label.format(mse_2009, ssim_2009))
grid[1].set_title('2010_2011')
#grid.axes_llc.set_yticks([0, 200, 400,600,])
plt.savefig("ssim_ropho.png", bbox_inches = 'tight', pad_inches = 0,dpi=100)

grid[2].imshow(img3, cmap=plt.cm.gray, vmin=0, vmax=1)
grid[2].set_xlabel(label.format(mse_2019, ssim_2019))
grid[2].set_title('2010_2019')
#grid.axes_llc.set_yticks([0, 200, 400,600,])
plt.savefig("ssim_ropho.png", bbox_inches = 'tight', pad_inches = 0,dpi=100)

plt.show()
```

```
from skimage.measure import compare_ssim, compare_psnr

mse_score=[]
ssim_score=[]
ps_score =[]

gt_img = plt.imread('/content/images/44041162_CR_20100510_001.jpg')
gt_img = rgb2gray(gt_img)
gt = img_as_float(gt_img)

def mse(x, y):
    return np.linalg.norm(x - y)

for item in sorted(glob.glob('/content/images/*.jpg')):
  tar_img = plt.imread(item)
  tar_gray= rgb2gray(tar_img)
  img = img_as_float(tar_gray)

  f,e = os.path.splitext(item)
  ms = mse(gt, img)
  ssim_none = compare_ssim(gt, img, data_range=img.max() - img.min())
  ps = compare_psnr(gt,img)
  mse_score.append([f,ms])
  ssim_score.append([f,ssim_none])
  ps_score.append([f,ps])
```

```
print(ssim_score)
```

```
import pandas as pd
m = pd.DataFrame(mse_score,columns = ['id', 'mse'])
ss = pd.DataFrame(ssim_score, columns = ['id', 'ssim'])
ps = pd.DataFrame(ps_score, columns = ['id', 'psnr'])
df = pd.merge(m,ss)
df = pd.merge(df,ps)
df.head()
```

```
id_date = df['id'].str.split('_',expand=True)
id_date.columns=['dir','type','date','num']
id_date2 = id_date['dir'].str.split('/', expand=True)
id_date2.columns=['null', 'cd','cd2','patient_id']
```

```
df2 = pd.concat([df,id_date2,id_date],axis=1)
df2.head()
```

```
id_group = df2.groupby('patient_id')
```

```
print(id_group.mean())
```
